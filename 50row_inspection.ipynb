```python
import pandas as pd
import matplotlib as plt
import sqlite3
```

```python
db_path = "raw/armd.sqlite"
```

```python
table_name = "implied_susceptibility_rules"
def extract_partial_data(db_path_internal = "raw/armd.sqlite", table_name_interal = table_name):
    with sqlite3.connect(db_path_internal) as conn:
        df = pd.read_sql_query(f"SELECT * FROM {table_name_interal} LIMIT 50", conn)
        column_names = df.columns.tolist()
    print(column_names)
    return(df)
    
```

```python
extract_partial_data(db_path, table_name)
```

# Interpretation
"This table captures cases where susceptibility to one antibiotic can imply susceptibility or resistance to another, based on established microbiological and pharmacological principles... The table is designed to enhance the interpretability of susceptibility data by incorporating implied relationships between antibiotics, which can be critical for guiding clinical decision-making and understanding resistance patterns." Ref: https://www.nature.com/articles/s41597-025-05649-7#Sec8
